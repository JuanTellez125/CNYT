## Exercise 2: Visualizing Complex Operations
1. Write a Python function to plot a complex number on the complex plane. Use it to plot `(3 + 4j)` and its conjugate.
2. Extend the function to show addition and multiplication of two complex numbers graphically.

```python
#1
import matplotlib.pyplot as plt

def plot_complex(z, label=None, color='b'):
    plt.scatter(z.real, z.imag, color=color, label=label)
    plt.text(z.real, z.imag, f'{z}', fontsize=12, ha='right')

# Números complejos
z1 = 3 + 4j
z1_conj = z1.conjugate()

plt.figure(figsize=(6,6))
plot_complex(z1, label='3 + 4j', color='blue')
plot_complex(z1_conj, label='Conjugado (3 - 4j)', color='red')

plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginaria')
plt.title('Plano Complejo')

#2.
def plot_operations(z1, z2):
    plt.figure(figsize=(7,7))
    # Plot original numbers
    plot_complex(z1, label=f'{z1}', color='blue')
    plot_complex(z2, label=f'{z2}', color='green')
    # Plot suma
    suma = z1 + z2
    plot_complex(suma, label=f'Suma: {suma}', color='orange')
    # Plot multiplicación
    mult = z1 * z2
    plot_complex(mult, label=f'Mult: {mult}', color='purple')
    # Ejes y detalles
    plt.axhline(0, color='gray', linewidth=0.5)
    plt.axvline(0, color='gray', linewidth=0.5)
    plt.xlabel('Parte Real')
    plt.ylabel('Parte Imaginaria')
    plt.title('Operaciones con Números Complejos')
    plt.legend()
    plt.grid(True)
    plt.show()

z2 = 1 - 2j

plot_operations(z1, z2)
```

## Exercise 3: Exploring the Mandelbrot Set
1. Modify the provided Mandelbrot set code to change its zoom level and center point. Observe how the fractal pattern changes.
2. Experiment with different values of `max_iter` and observe the effect on the fractal's detail and computation time.

```python
#1
import numpy as np
def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

center_x, center_y = -0.5, 0.0  # Cambia estos valores para mover el centro
zoom = 0.5  # <1 para acercar, >1 para alejar

pixels = 800
max_iter = 30 

# Calcula los límites usando zoom y centro
x_width = 3 * zoom
y_height = 3 * zoom

x_min = center_x - x_width / 2
x_max = center_x + x_width / 2
y_min = center_y - y_height / 2
y_max = center_y + y_height / 2

x = np.linspace(x_min, x_max, pixels)
y = np.linspace(y_min, y_max, pixels)
xx, yy = np.meshgrid(x, y)
c = xx + yy*1j

mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

plt.imshow(mandelbrot_set, extent=[x_min, x_max, y_min, y_max])
plt.colorbar()
plt.title(f"Mandelbrot Set (center=({center_x},{center_y}), zoom={zoom})")
plt.show()

#2. 
"""Entre mas iteraciones haya, mas detalles se pueden observar en el conjunto de Mandelbrot
y mas tiempo tarda en generarse la imagen, mientras que, menos interaciones haya, menos detalles se veran y 
menos tiempo se tardara."""


```

## Exercise 4: Creating a Julia Set
1. Implement a Julia set generator. Use a constant like `-0.4 + 0.6j` for the iteration `z = z*z + constant`.
2. Explore how changing the constant changes the pattern of the Julia set.

```python
#1.
"""Para este ejercicio vamos a hacer un ajuste al codigo usado en el anterior punto"""

def julia(z, constant, max_iter):
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + constant
    return max_iter

# Parámetros
constant1 = -0.4 + 0.6j #Este es el ajuste
pixels = 800
max_iter = 30

x = np.linspace(-1.5, 1.5, pixels)
y = np.linspace(-1.5, 1.5, pixels)
xx, yy = np.meshgrid(x, y)
z = xx + yy*1j

julia_set1 = np.array([julia(zi, constant1, max_iter) for zi in z.ravel()]).reshape(z.shape)

plt.imshow(julia_set1, extent=[-1.5, 1.5, -1.5, 1.5])
plt.colorbar()
plt.title(f"Julia Set (constant={constant1})")
plt.show()

#2. 
"""Al cambiar la constante, la forma y los detalles del conjunto de Julia cambian significativamente. Algunos
valores generan una conexion entre el area y otros patrones mas dispersos y desconectados."""

constant2 = 0.355 + 0.355j
constant3 = -0.70176 - 0.3842j
constant4 = 0.8 + 0.156j 

julia_set2 = np.array([julia(zi, constant2, max_iter) for zi in z.ravel()]).reshape(z.shape)

plt.imshow(julia_set2, extent=[-1.5, 1.5, -1.5, 1.5])
plt.colorbar()
plt.title(f"Julia Set (constant={constant2})")
plt.show()

julia_set3 = np.array([julia(zi, constant3, max_iter) for zi in z.ravel()]).reshape(z.shape)

plt.imshow(julia_set3, extent=[-1.5, 1.5, -1.5, 1.5])
plt.colorbar()
plt.title(f"Julia Set (constant={constant3})")
plt.show()

julia_set4 = np.array([julia(zi, constant4, max_iter) for zi in z.ravel()]).reshape(z.shape)

plt.imshow(julia_set4, extent=[-1.5, 1.5, -1.5, 1.5])
plt.colorbar()
plt.title(f"Julia Set (constant={constant4})")
plt.show()
```

## Exercise 5: Complex Number Properties
1. Prove that the absolute value of a product of two complex numbers is the product of their absolute values.
2. Show that the conjugate of a sum of two complex numbers is the sum of their conjugates.

```python
#1.
c1 = -0.4 + 0.6j
c2 = 0.355 + 0.355j
producto_valorabs = abs(c1 * c2)
print("Valor absoluto del producto :", producto_valorabs)
valorabs_producto = abs(c1) * abs(c2)
print("Producto de los valores absolutos:", valorabs_producto)

#2.
conjugado_sum = (c1 + c2).conjugate()
print("Conjugado de la suma:", conjugado_sum)
sum_conjugados = c1.conjugate() + c2.conjugate()
print("Suma de los conjugados:", sum_conjugados)
```
