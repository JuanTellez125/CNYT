```python

import numpy as np

# Se definen dos vectores de números complejos
v1 = np.array([[4+2j], [3-5j]])
v2 = np.array([[5-1j], [2+6j]])

#Se definen dos matrices de números complejos
m1 = np.array([[1+1j, 2], [3, 4+2j]])
m2 = np.array([[0, 1j], [2, 3]])

# 1. Adicion de vectores complejos 
sum_v = v1 + v2

# 2. inverso aditivo de vectores complejos
neg_v1 = -v1

# 3. multiplicación de un escalar por un vector complejo
scalar = 2 + 1j
mult_v1 = v1 * scalar

# 4. Suma de matrices complejas
sum_m = m1 + m2

# 5. Inverso aditivo de una matriz compleja
neg_m1 = -m1

# 6. Multiplicación de un escalar por una matriz compleja
mult_m1 = m1 * scalar # mismo escalar que en el vector

# 7. Transpuesta de un vector complejo y matriz compleja
transpose_v1 = v1.T
transpose_m1 = m1.T

# 8. Conjugada de un vector complejo y matriz compleja
conjugate_v1 = np.conjugate(v1)
conjugate_m2 = np.conjugate(m2)

# 9. Adjunta (conjugada transpuesta) de un vector complejo y matriz compleja
adjunct_v1 = v1.T.conj()
adjunct_m1 = m1.T.conj()

# 10. Producto de dos matrices complejas de tamaños compatibles
product_m = np.dot(m1, m2)

# 11. Funcion para calcular la "accion" de una matriz sobre un vector
accion = np.dot(m1, v1)

# 12. Producto interno de dos vectores complejos
dot_product_v = np.vdot(v1, v2)

# 13. Norma de un vector complejo
norm_v1 = np.linalg.norm(v1)

# 14. Distancia entre dos vectores complejos
distance_v = np.linalg.norm(v1 - v2)

# 15. Valores propios y vectores propios de una matriz compleja
eigenvalues, eigenvectors = np.linalg.eig(m1)

# 16. Revisar si una matriz es unitaria
is_unitary = np.allclose(np.eye(m1.shape[0]), np.dot(m1, m1.T.conj()))

# 17. Revisar si una matriz es hermitiana
is_hermitian = np.array_equal(m1, m1.T.conj())

# 18. Producto tensor de dos vectores o matrices complejas
tensor_product_v = np.kron(v1, v2)
tensor_product_m = np.kron(m1, m2)


print("1. Adicion de vectores complejos:\n",sum_v)
print("2. Inverso  (Aditivo) de un vector complejo\n",neg_v1)
print("3. Multiplicacion de un escalar por un numero complejo\n",mult_v1)
print("7. Transpuesta de un vector complejo\n",transpose_v1)
print("7. Transpuesta de una matriz compleja\n",transpose_m1)
print("8. Conjugada de un vector complejo\n",conjugate_v1)
print("8. Conjugada de una matriz compleja\n",conjugate_m2)
print("9. Adjunta de un vector complejo\n",adjunct_v1)
print("9. Adjunta de una matriz compleja\n",adjunct_m1)
print("10. Producto de dos matrices complejas\n",np.dot(m1,m2))
print("11. Accion de una matriz sobre un vector\n",accion)
print("12. Producto interno de dos vectores complejos\n",dot_product_v)
print("13. Norma de un vector complejo\n",norm_v1)
print("14. Distancia entre dos vectores complejos\n",distance_v)
print("15. Valores propios de una matriz compleja\n",eigenvalues)
print("15. Vectores propios de una matriz compleja\n",eigenvectors)
print("16. La matriz es unitaria?\n",is_unitary)
print("17. La matriz es hermitiana?\n",is_hermitian)
print("18. Producto tensor de dos vectores complejos\n",tensor_product_v)
print("18. Producto tensor de dos matrices complejas\n",tensor_product_m)


```
